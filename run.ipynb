```python
# SFT 数据质量评估与清洗工具 - 运行入口
import os
import logging
from config import Config
from pipeline import SFTQualityPipeline

# 配置日志格式
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# ==================== 用户配置区域 ====================
# 1. 如需启用 LLM 混合评分，请在此处设置 API Key（推荐使用环境变量）
os.environ["DEEPSEEK_API_KEY"] = "API Key"

# 2. 基础配置（采样数量在 run 方法中指定，此处可留空或设为 None）
config = Config(
    n_samples=None,                     # 这里不指定，由 run 方法传入
    enable_llm_scoring=True,           # 改为 True 启用混合评分（需先配置 API Key）
    high_score_threshold=8,             # 高质量分数阈值
    low_score_threshold=6,              # 低质量分数阈值
    output_dir="output_sft"             # 输出目录
)

# 3. 运行评估流水线
pipeline = SFTQualityPipeline(config)

# 数据目录（请根据实际情况修改）
data_path = "./SFT数据集"               # 可以是相对路径或绝对路径
sample_size = 20                       # 采样数量，设为 None 表示处理全部数据

result = pipeline.run(data_dir=data_path, verbose=True, n_samples=sample_size)

# ==================== 输出结果 ====================
print("\n" + "="*50)
print(f"处理完成！高质量样本数: {len(result['high_quality'])}")
print(f"详细结果请查看: {config.output_dir}/ 目录")
```
